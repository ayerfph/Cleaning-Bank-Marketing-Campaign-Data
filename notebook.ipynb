### Import and setting up the DataFrames / Tables

```python
# Import
import pandas as pd
import numpy as np
```

```python
# Reading and splitting the data
marketing = pd.read_csv("bank_marketing.csv")

# Split into three tables
client = marketing[["client_id","age","job","marital","education","credit_default","mortgage"]]
campaign = marketing[["client_id","number_contacts","month","day","contact_duration","previous_campaign_contacts","previous_outcome","campaign_outcome"]]
economics = marketing[["client_id","cons_price_idx","euribor_three_months"]]
```

### Editing Client Dataset

```python
# Cleaning education column
client["education"] = client["education"].str.replace(".","_")
client["education"] = client["education"].replace("unknown", np.NaN)
```

```python
# Clean job column
client["job"] = client["job"].str.replace(".","_")
```

```python
# Clean and convert client columns to bool data type
for col in["credit_default","mortgage"]:
    client[col] = client[col].map({"yes":1,"no":0,"unknown":0})
    client[col] = client[col].astype(bool)
```

```python
client.head(5)
```

### Editing Campaign Dataset

```python
# Change campaign_outcomes to binary values
campaign["campaign_outcome"] = campaign["campaign_outcome"].map({"yes": 1, "no": 0})
```

```python
# Convert previous_outcome to binary values
campaign["previous_outcome"] = campaign["previous_outcome"].map({"sucess":1, "failure":0, "nonexistent":0})
```

```python
# Add year column
campaign["year"] = "2022"
```

```python
# Convert day to string
campaign["day"] = campaign["day"].astype(str)
```

```python
# Add last_contact_date column
campaign["last_contact_date"] = campaign["year"] + "-" + campaign["month"] + "-" + campaign["day"]
```

```python
# Convert to datetime
campaign["last_contact_date"] = pd.to_datetime(campaign["last_contact_date"], format="%Y-%b-%d")
```

```python
# Clean and convert outcome columns to bool
for col in ["campaign_outcome","previous_outcome"]:
    campaign[col] = campaign[col].astype(bool)
```

```python
# Drop Unneccessary columns
campaign.drop(columns=["month","day","year"], inplace=True)
```

### Convert all tables into csv files

```python
client.to_csv("client.csv", index=False)
campaign.to_csv("campaign.csv", index=False)
economics.to_csv("economics.csv", index=False)
```
